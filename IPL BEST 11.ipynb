# Cleaning the data 

In the abive data players can be initially segregatted on the basis on capped and uncapped players
  1.Capped :- Batters,Bowleres,Allrounders,wicketkeepers
  2.uncapped :- Batters,Bowleres,Allrounders,wicketkeepers
  The incapped players are playing the IPl first time therefor we have used domestic t20 data

```python
# # segragating data - capped batters
# we have kept the batting parameters to restricts the analysis for batter specific tendencies
```

```python
batters = data.loc[(data['Player_Type'] == 'Batter')]
batters_new = batters.loc[(batters['Capped'] == 1)]
capped_batter = batters_new[['Player Name','Team','Nationality','Matches_Played','Runs','Average','Strike_Rate']]
```

```python
# The seggregation of batters based on capped data based on batters who have ipl match experience
```

```python
capped_batter.head()
```

```python
# segragating data capped bowlers
```

```python
bowlers = data.loc[(data['Player_Type'] == 'Bowler ')]
bowlers_new = bowlers.loc[(bowlers['Capped'] == 1)]
capped_bowlers = bowlers_new[['Player Name','Team','Nationality','Matches_Played','Wickets','Bowling_average','Economy','Bowling_Strike_Rate']]
```

```python
capped_bowlers.head()
```

```python
#seggregating data for keepers
```

```python
keepers = data.loc[(data['Player_Type'] == 'Keeper')]
keepers_new = keepers.loc[(keepers['Capped'] == 1)]
capped_keppers = keepers_new[['Player Name','Team','Nationality','Matches_Played','Runs','Average','Strike_Rate','Catches','Run_outs','Stumps']]
```

```python
capped_keppers.head()
```

```python
#seggregating data for Allrounders
```

```python
allrounders = data.loc[(data['Player_Type'] == 'Allrounder')]
allrounders_new = allrounders.loc[(allrounders['Capped'] == 1)]
capped_allrounders = allrounders_new[['Player Name','Team','Nationality','Matches_Played','Runs','Average','Strike_Rate','Wickets','Bowling_average','Economy','Bowling_Strike_Rate']]
```

```python
capped_allrounders.head()
```

```python
# cleaning the data by filling null value with 0
```

```python
capped_batter = capped_batter.fillna(0)
capped_bowlers = capped_bowlers.fillna(0)
capped_keppers = capped_keppers.fillna(0)
capped_allrounders = capped_allrounders.fillna(0)
```

```python
# checking the null value 
print(capped_batter.isna().sum())
print(capped_bowlers.isna().sum())
print(capped_keppers.isna().sum())
print(capped_allrounders.isna().sum())
```
